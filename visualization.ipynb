```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
n_perfomance = pd.read_csv('n_perfomance.csv', delimiter=';')
n_perfomance.head(5)
```

```python
fig = make_subplots(rows=2, cols=1, horizontal_spacing=0.15)
fig.add_trace(
    go.Scatter(x=n_perfomance.n, y=n_perfomance.area),
    row=1, col=1
)
fig.update_xaxes(title='n',
                 row=1, col=1)
fig.update_yaxes(title='y_pred',
                 row=1, col=1)

fig.add_trace(
    go.Scatter(x=n_perfomance.n, y=n_perfomance['diff']),
    row=2, col=1
)
fig.update_xaxes(title='n',
                 row=2, col=1)
fig.update_yaxes(title='residuals',
                 row=2, col=1)

fig.update_layout(title='Зависимость качества алгоритма от кол-ва поставленных точек',
                  showlegend=False,
                  width=700, height=1000)
```

- При увеличении плотности точек точность предсказанной площади растет.
- При увеличении плотности точек точность предсказанной площади растет => уменьшается и разница с правильным значением.

```python
scale_perfomance = pd.read_csv('scale_perfomance.csv', delimiter=';')
scale_perfomance.head(5)
```

```python
fig = make_subplots(rows=2, cols=1, horizontal_spacing=0.15)
fig.add_trace(
    go.Scatter(x=scale_perfomance.scale, y=scale_perfomance.area),
    row=1, col=1
)
fig.update_xaxes(title='scale',
                 row=1, col=1)
fig.update_yaxes(title='y_pred',
                 row=1, col=1)

fig.add_trace(
    go.Scatter(x=scale_perfomance.scale, y=scale_perfomance['diff']),
    row=2, col=1
)
fig.update_xaxes(title='scale',
                 row=2, col=1)
fig.update_yaxes(title='residuals',
                 row=2, col=1)

fig.update_layout(title='Зависимость качества алгоритма от масштаба',
                  showlegend=False,
                  width=700, height=1000)
```

- При увеличении масштаба, плотность расставленных точек уменьшается => уменьшается точность.
- Соотвественно и увеличивается разница с правильным ответом. 

ВЫВОД: При использованиит метода Монте-Карла по возможности стоит выбирать большее кол-во точек и наименьший возможный масштаб.
